# **2. Thu thập dữ liệu** 

Ở bước thu thập dữ liệu, chúng em sẽ thu thập dữ liệu qua API với các tham số như đặc trưng, quốc gia, năm.

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
INDICATOR_CODES = ['SP.POP.TOTL', # Tổng dân số
                   'SP.POP.TOTL.FE.IN', # Dân số nữ
                   'SP.POP.TOTL.MA.IN', # Dân số nam
                   'SP.DYN.CBRT.IN', # Tỉ lệ sinh(Trên 1000 dân)
                   'SP.DYN.CDRT.IN', # Tỉ lệ tử(Trên 1000 dân)
                   'NY.GDP.MKTP.CD', # GDP(USD)
                   'SL.UEM.TOTL.ZS', # Tỉ lệ thất nghiệp(%)
                   'SP.URB.TOTL.IN.ZS', # Tỉ lệ dân số đô thị(%)
                   'NE.EXP.GNFS.ZS', # Xuất khẩu hàng hóa và dịch vụ(%GPD)
                   'NE.IMP.GNFS.ZS', # Nhập khẩu hàng hóa và dịch vụ(%GPD)
                   'NY.GDP.MKTP.KD.ZG', #Tốc độ tăng trưởng GPD(%)
                   'SP.URB.GROW', # Tốc độ tăng trưởng dân số đô thị hằng năm(%)
                   'SP.DYN.TFRT.IN', # Tổng tỉ suất sinh(Số con trung bình của 1 phụ nữ)
                   'SL.AGR.EMPL.ZS', # Tỉ lệ lao động ở ngành nông nghiệp(%)
                   'SL.IND.EMPL.ZS', # Tỉ lệ lao động ở ngành công nghiệp(%)
                   'SL.SRV.EMPL.ZS',  # Tỉ lệ lao động ở ngành dịch vụ(%)
                   'NV.AGR.TOTL.ZS', # Tỉ trọng ngành nông nghiệp(%GDP)
                   'NV.IND.TOTL.ZS', # Tỉ trọng ngành công nghiệp(%GDP)
                   'NV.SRV.TOTL.ZS', # Tỉ trọng ngành dịch vụ(%GDP)
                   ]
```

```python
COUNTRY_CODES = ['AUS', 'AFG', 'BGD', 'BTN', 'BRA', 'CHL', 'CAN', 'CHN', 'CUB', 'DEU', 'DNK', 'EGY', 'ETH', 'GBR',
                  'IDN', 'IND', 'JPN', 'KHM', 'LBY', 'MMR', 'MEX', 'SGP', 'PHL', 'RUS', 'THA', 'USA', 'VNM', 'ZWE']
```

```python
def loadData(country_code, start, end, format = "json", per_page = 100):

    records = {}
    for indicator_code in INDICATOR_CODES:
        url = f"https://api.worldbank.org/v2/country/{country_code}/indicator/{indicator_code}?date={start}:{end}&format=json&per_page={per_page}"
        response = requests.get(url) # Lấy về kết quả sau khi gọi api

        if response.status_code == 200:

            data = response.json() # Chuyển kết quả về dạng json

            # Xử lí data để thu được các trường thông tin mong muốn
            count = data[0]['total']

            for i in range(count):
                item = data[1][i]
                feature = item['indicator']['value']
                country = item['country']['value']
                year = item['date']
                value = item['value']

                if year not in records:
                    records[year] = {"Year": year}
                
                records[year]['Country'] = country

                records[year][feature] = value
            

        df = pd.DataFrame(records.values())

    return df
        
    
```

```python
raw_data = pd.DataFrame()
for country in COUNTRY_CODES:
    print(country)

    df = loadData(country, 1980, 2020)
    raw_data = pd.concat([raw_data, df], ignore_index=True)
```

```python
raw_data.head()
raw_data.to_csv('Data.csv', index=False)
```

- Vì đang có khá nhiều nước nên khi trực quan sẽ gây khó khăn cho người xem. Do đó chúng em quyết định sẽ phân loại các nước theo các nhóm `Developed`, `Developing`, `Least developed`.
- Do đó, chúng em sẽ điền các giá trị rỗng bằng giá trị trung bình của nhóm nước đó. Vì thường các nước cùng nhóm sẽ có các chỉ số gần nhau.

```python
developed_countries = ['Australia', 'Canada', 'Germany', 'Denmark', 'United Kingdom', 'Japan', 'Singapore', 'United States']
undeveloped_countries = ["Cambodia", "Myanmar", "Zimbabwe"]
# Đánh chỉ số 0 cho các nước phát triển, 1 cho các nước đang phát triển, 2 cho các nước chưa phát triển
raw_data['Development_Status'] = raw_data['Country'].apply(lambda x: 'Developed' if x in developed_countries else ('Least developed' if x in undeveloped_countries else 'Developing'))
```

```python
# Điền giá trị rỗng bằng trung bình của nhóm
numeric_columns = raw_data.select_dtypes(include=['number']).columns
raw_data[numeric_columns] = raw_data.groupby("Development_Status")[numeric_columns].transform(lambda x: x.fillna(x.mean()))
raw_data.to_csv("processed_data.csv", index=False)
```

# **4. Phân tích thông tin cơ bản bộ dữ liệu**

## **4.1 Giới thiệu bộ dữ liệu:**

Bộ dữ liệu được lấy từ **WorldBank Open Data**, gồm nhiều record của nhiều quốc gia trong giai đoạn 1980 đến 2020.

## **4.2 Cấu trúc và ý nghĩa trường dữ liệu:**


au quá trình tiền xử lý dữ liệu, bộ dữ liệu bao gồm **1148** dòng, với **21** trường dữ liệu như sau:  

- **`Year`** (*Năm*) – Thể hiện năm thu thập dữ liệu.  

- **`Country`** (*Quốc gia*) – Tên quốc gia tương ứng với dữ liệu thu thập.  

- **`Population, total`** (*Tổng dân số*) – Số lượng dân cư của quốc gia trong năm.  

- **`Population, female`** (*Dân số nữ*) – Tổng số nữ giới trong dân số của quốc gia.  

- **`Population, male`** (*Dân số nam*) – Tổng số nam giới trong dân số của quốc gia.  

- **`Birth rate, crude (per 1,000 people)`** (*Tỷ suất sinh thô*) – Số trẻ sinh ra sống trung bình trên 1.000 người dân trong năm.  

- **`Death rate, crude (per 1,000 people)`** (*Tỷ suất tử vong thô*) – Số ca tử vong trung bình trên 1.000 người dân trong năm.  

- **`GDP (current US$)`** (*Tổng sản phẩm quốc nội (GDP)*) – Giá trị GDP danh nghĩa của quốc gia, tính theo USD.  

- **`Unemployment, total (% of total labor force) (modeled ILO estimate)`** (*Tỷ lệ thất nghiệp*) – Tỷ lệ thất nghiệp tính trên tổng lực lượng lao động, dựa trên ước tính của ILO.  

- **`Urban population (% of total population)`** (*Tỷ lệ dân số đô thị*) – Phần trăm dân số sống ở khu vực đô thị so với tổng dân số quốc gia.  

- **`Exports of goods and services (% of GDP)`** (*Xuất khẩu hàng hóa và dịch vụ*) – Tổng giá trị hàng hóa và dịch vụ xuất khẩu, tính theo phần trăm GDP.  

- **`Imports of goods and services (% of GDP)`** (*Nhập khẩu hàng hóa và dịch vụ*) – Tổng giá trị hàng hóa và dịch vụ nhập khẩu, tính theo phần trăm GDP.  

- **`GDP growth (annual %)`** (*Tăng trưởng GDP hằng năm*) – Mức tăng trưởng GDP của quốc gia so với năm trước, tính theo phần trăm.  

- **`Urban population growth (annual %)`** (*Tăng trưởng dân số đô thị hằng năm*) – Tốc độ tăng trưởng dân số đô thị của quốc gia theo phần trăm mỗi năm.  

- **`Fertility rate, total (births per woman)`** (*Tổng tỷ suất sinh*) – Số con trung bình mà một phụ nữ có thể sinh trong suốt cuộc đời.  

- **`Employment in agriculture (% of total employment) (modeled ILO estimate)`** (*Tỷ lệ lao động trong nông nghiệp*) – Phần trăm tổng số lao động làm việc trong lĩnh vực nông nghiệp, dựa trên ước tính của ILO.  

- **`Employment in industry (% of total employment) (modeled ILO estimate)`** (*Tỷ lệ lao động trong công nghiệp*) – Phần trăm tổng số lao động làm việc trong lĩnh vực công nghiệp, bao gồm xây dựng và sản xuất.  

- **`Employment in services (% of total employment) (modeled ILO estimate)`** (*Tỷ lệ lao động trong dịch vụ*) – Phần trăm tổng số lao động làm việc trong lĩnh vực dịch vụ.  

- **`Agriculture, forestry, and fishing, value added (% of GDP)`** (*Giá trị gia tăng ngành nông, lâm, ngư nghiệp*) – Tỷ lệ đóng góp của ngành nông, lâm, ngư nghiệp vào GDP quốc gia.  

- **`Industry (including construction), value added (% of GDP)`** (*Giá trị gia tăng ngành công nghiệp và xây dựng*) – Tỷ lệ đóng góp của ngành công nghiệp và xây dựng vào GDP quốc gia.  

- **`Services, value added (% of GDP)`** (*Giá trị gia tăng ngành dịch vụ*) – Tỷ lệ đóng góp của ngành dịch vụ vào GDP quốc gia.  

- **`Development_Status`** (*Trạng thái phát triển*) – Phân loại quốc gia theo mức độ phát triển, bao gồm "Developed" (*Phát triển*), "Developing" (*Đang phát triển*), và "Least developed" (*Kém phát triển nhất*).

### **Câu 2: Tình hình dân số của các nhóm nước**

```python
year_start = group_data['Year'].min()
year_end = group_data['Year'].max()

# Lấy dữ liệu cho từng năm
data_start = group_data[group_data['Year'] == year_start]['Population, total']
data_end = group_data[group_data['Year'] == year_end]['Population, total']

total_start = data_start.sum()
total_end = data_end.sum()

# Vẽ hai biểu đồ tròn
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
fig.subplots_adjust(wspace=0.4) 


# Biểu đồ mốc đầu
axes[0].pie(data_start, labels=data_start.index, autopct='%1.1f%%', startangle=140)
axes[0].set_title(f'Dân số theo nhóm nước ({year_start})')
axes[0].text(0, 0, f"{total_start/1e9:.2f}B", ha='center', va='center', fontsize=14, fontweight='bold')

# Biểu đồ mốc cuối
axes[1].pie(data_end, labels=data_end.index, autopct='%1.1f%%', startangle=140)
axes[1].set_title(f'Dân số theo nhóm nước ({year_end})')
axes[1].text(0, 0, f"{total_end/1e9:.2f}B", ha='center', va='center', fontsize=14, fontweight='bold')

plt.show()
```

Để có cái nhìn tổng quát về quy mô dân số ở 2 mốc đầu cuối, chúng em dùng 2 biểu đồ tròn để trực quan, thể hiện tỉ lệ đóng góp vào tổng dân số ở các nhóm nước, quy mô tổng dân số.

**Phân tích:**
- Dân số tăng nhiều trong khoảng 40 năm(hơn 50%)
- Các nước đang phát triển đông dân, có tốc độ gia tăng dân số nhanh.
- Các nước phát triển có tốc độ gia tăng dân số chậm

Để nhìn rõ tốc độ gia tăng dân số, tổng số dân của các nhóm nước, em sẽ dùng 2 biểu đồ đường để trực quan tổng số dân và tỉ lệ số con trên mỗi người phụ nữ ở các nhóm quốc gia theo thời gian

```python
# Biểu diễn tổng dân số của các nhóm quốc gia theo năm
plt.figure(figsize=(12, 6))
sns.lineplot(data=group_data, x='Year', y='Population, total', hue='Development_Status')

plt.title('Dân số của các nhóm quốc gia')
plt.xlabel('Năm')
plt.ylabel('Dân số')
plt.legend(title='Nhóm nước')
plt.show()
```

### **Câu 4: Tỷ lệ đô thị hóa có liên quan gì đến tỷ lệ thất nghiệp.**

Kết luận:

### **Câu 5: Cơ cấu lao động theo đầu người trong các ngành thay đổi như thế nào ở từng nhóm quốc gia và có liên quan gì đến tỷ lệ thất nghiệp?**

```python


development_groups = {
    "Developed" : 'Nước phát triển',
    "Developing" : 'Nước đang phát triển',
    "Least developed" : 'Nước chậm phát triển'
}

# Tạo figure với 3 subplots
fig, axes = plt.subplots(3, 1, figsize=(12, 18), sharex=False)

# Lặp qua từng nhóm phát triển để vẽ biểu đồ
for i, (status, title) in enumerate(development_groups.items()):
    subset = group_df_filtered[group_df_filtered['Development_Status'] == status]

    # Chuyển đổi dữ liệu sang dạng phù hợp cho biểu đồ cột nhóm
    subset_melted = subset.melt(id_vars=['Year'], 
                                value_vars=['Employment in agriculture (% of total employment) (modeled ILO estimate)',
                                            'Employment in industry (% of total employment) (modeled ILO estimate)',
                                            'Employment in services (% of total employment) (modeled ILO estimate)'],
                                var_name='Ngành nghề', 
                                value_name='Tỷ lệ lao động')

    # Đổi tên ngành nghề để dễ đọc
    subset_melted['Ngành nghề'] = subset_melted['Ngành nghề'].replace({
        'Employment in agriculture (% of total employment) (modeled ILO estimate)': 'Nông nghiệp',
        'Employment in industry (% of total employment) (modeled ILO estimate)': 'Công nghiệp',
        'Employment in services (% of total employment) (modeled ILO estimate)': 'Dịch vụ'
    })

    # Vẽ biểu đồ cột nhóm
    sns.barplot(data=subset_melted, x='Year', y='Tỷ lệ lao động', hue='Ngành nghề', ax=axes[i])
    # Cài đặt tiêu đề và nhãn
    axes[i].set_title(f'Cơ cấu lao động ({title})')
    axes[i].set_ylabel('Tỷ lệ lao động (%)')
    axes[i].set_xlabel('Năm')
    axes[i].legend(title='Ngành nghề')
    axes[i].grid(axis='y')
# Hiển thị biểu đồ
plt.show()
```

Phân tích:
- Cơ cấu lao động ở nước phát triển:
+ Ngành dịch vụ (màu xanh lá) chiếm tỷ trọng áp đảo (> 60%) và tiếp tục tăng dần theo thời gian.
+ Ngành công nghiệp (màu cam) duy trì ở mức trung bình (~ 20-30%), ít biến động.
+ Ngành nông nghiệp (màu xanh dương) gần như biến mất (< 5%), cho thấy nông nghiệp không còn là ngành trọng yếu về lao động.
+ Tỉ lệ này vẫn không thay đổi nhiều trong vòng 40 năm.
- Cơ cấu lao động ở nước đang phát triển:
+ Ngành nông nghiệp (màu xanh dương) vẫn chiếm tỷ lệ đáng kể nhưng đang giảm dần qua các năm.
+ Ngành công nghiệp (màu cam) có xu hướng ổn định ở mức trung bình (~ 20-30%).
+ Ngành dịch vụ (màu xanh lá) ngày càng tăng, cho thấy sự chuyển dịch lao động sang lĩnh vực này.
+ Các nước đang phát triển đang trong quá trình chuyển đổi từ nền kinh tế nông nghiệp sang công nghiệp và dịch vụ, nhưng chưa hoàn toàn chuyển đổi mạnh như các nước phát triển.
- Cơ cấu lao động ở nước chậm phát triển:
+ Ngành nông nghiệp (màu xanh dương) vẫn chiếm tỷ trọng lớn nhất (> 50%) nhưng có xu hướng giảm dần.
+ Ngành công nghiệp (màu cam) vẫn duy trì ở mức thấp, ít thay đổi.
+ Ngành dịch vụ (màu xanh lá) bắt đầu tăng nhưng chưa đáng kể.
+ Ở các nước chậm phát triển, nông nghiệp vẫn là ngành chủ đạo, dù có xu hướng giảm dần. Công nghiệp chưa phát triển mạnh và dịch vụ vẫn còn ở mức thấp.

### **Câu 9: Các quốc gia đang phát triển nào có tiềm năng trở thành quốc gia phát triển trong tương lai.**

```python

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors



# Chọn các cột quan trọng để phân tích
features = ["GDP (current US$)", "Urban population (% of total population)", 
            "Unemployment, total (% of total labor force) (modeled ILO estimate)", 
            "GDP growth (annual %)", "Employment in agriculture (% of total employment) (modeled ILO estimate)", "Employment in industry (% of total employment) (modeled ILO estimate)"
            ,"Employment in services (% of total employment) (modeled ILO estimate)"]

temp = df[df['Year'] == 2020]
# Lọc dữ liệu theo nhóm phát triển
group_0 = temp[temp["Development_Status"] == 'Developed']
group_1 = temp[temp["Development_Status"] == 'Developing']

# Chuẩn hóa dữ liệu
scaler = StandardScaler()
X_0 = scaler.fit_transform(group_0[features])
X_1 = scaler.transform(group_1[features])

# Áp dụng KNN để tìm khoảng cách từ nhóm 1 đến nhóm 0
knn = NearestNeighbors(n_neighbors=1, metric='euclidean')
knn.fit(X_0)
distances, indices = knn.kneighbors(X_1)

# Gán khoảng cách vào group_1 và sắp xếp theo khoảng cách tăng dần
group_1["distance_to_0"] = distances
top_candidates = group_1.sort_values(by="distance_to_0").head(5)  # Chọn top 10 nước gần nhóm 0 nhất

# Vẽ biểu đồ cột
plt.figure(figsize=(10, 6))
plt.barh(top_candidates["Country"][::-1], top_candidates["distance_to_0"][::-1], color='skyblue')

# Gắn nhãn
plt.xlabel("Khoảng cách đến nhóm quốc gia phát triển")
plt.ylabel("Quốc gia")
plt.title("Top 5 quốc gia có khả năng nâng hạng phát triển")
plt.grid(axis="x", linestyle="--", alpha=0.7)

# Hiển thị biểu đồ
plt.show()
```

- Phân tích và Nhận xét: 
    + Có thể thấy dựa vào việc lấy 2 nhóm đang phát triển và nhóm phát triển, ta sử dụng giải thuật NearestNeigbour của học máy để tính được khoảng cách từ các quốc gia trong nhóm đang phát triển đến nhóm phát triển, kết quả thu được 5 quốc gia có khoảng cách đến nhóm phát triển nhỏ nhất, tương ứng với đó là 5 quốc gia có khả năng nâng lên thành quốc gia phát triển.
    + Các quốc gia Nga, Chile, Brazil, Mexio, Libya là các quốc gia có khoảng cách ngắn nhất.
